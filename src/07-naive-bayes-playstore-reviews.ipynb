```python
import numpy as np
from sklearn.model_selection import RandomizedSearchCV
from sklearn.datasets import load_iris, fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report
```

# Lectura del dataframe

```python
df = pd.read_csv("https://raw.githubusercontent.com/4GeeksAcademy/naive-bayes-project-tutorial/main/playstore_reviews.csv")
```

package_name. Nombre de la aplicación móvil (categórico)

review. Comentario sobre la aplicación móvil (categórico)

polarity. Variable de clase (0 o 1), siendo 0 un comentario negativo y 1, positivo (categórico numérico)

```python
df
```

```python
df = df[["review","polarity"]]
```

```python
df["review"] = df["review"].str.strip().str.lower()
```

```python
# Dividimos el conjunto de datos en muestras de train y test
X = df["review"]
y = df["polarity"]

# Instancia de: from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.2,
                                                    random_state=18)
```

```python
# Vectorización con TfidfVectorizer
vectorizer = TfidfVectorizer()

X_train = vectorizer.fit_transform(X_train).toarray()
X_test = vectorizer.transform(X_test).toarray()
```

```python
#vec_model = CountVectorizer(stop_words = "english")
#X_train = vec_model.fit_transform(X_train).toarray()
#X_test = vec_model.transform(X_test).toarray()
```

```python
X_train
```

```python
# Entrenamiento del clasificador
model = MultinomialNB()

model.fit(X_train, y_train)
```

```python
X_train
```

```python
# Predicción
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)
```

```python
# Métricas
metrics = {"Accuracy Train: ": accuracy_score(y_train, y_pred_train),
           "Accuracy Test ": accuracy_score(y_test, y_pred_test)}

metrics
```

```python
hyperparams = {"alpha": np.linspace(0.01, 10.0, 100),
               "fit_prior": [True, False]}

# We initialize the random search
random_search = RandomizedSearchCV(model, hyperparams,
                                   n_iter=100,
                                   scoring="accuracy",
                                   cv=5,
                                   random_state=18)
random_search
```

```python
random_search.fit(X_train, y_train)

random_search.best_params_
```

```python
# Entrenamiento del clasificador
model = MultinomialNB(fit_prior= True, alpha=np.float64(0.01))
model.fit(X_train, y_train)

# Predicción
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)

# Métricas
metrics = {"Accuracy Train: ": accuracy_score(y_train, y_pred_train),
           "Accuracy Test ": accuracy_score(y_test, y_pred_test)}

metrics
```

demasiado overfitting pero nos indica que el mejor numero esta por debajo de 1

```python
hyperparams = {"alpha": np.linspace(0.60, 0.70, 1),
               "fit_prior": [True, False]}

# We initialize the random search
random_search = RandomizedSearchCV(model, hyperparams,
                                   n_iter=30,
                                   scoring="accuracy",
                                   cv=5,
                                   random_state=18)
random_search
random_search.fit(X_train, y_train)

random_search.best_params_
```

```python
# Entrenamiento del clasificador
model = MultinomialNB(fit_prior= True, alpha=np.float64(0.72))
model.fit(X_train, y_train)

# Predicción
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)

# Métricas
metrics = {"Accuracy Train: ": accuracy_score(y_train, y_pred_train),
           "Accuracy Test ": accuracy_score(y_test, y_pred_test)}

metrics
```

# Binomial


```python
# Entrenamiento del clasificador
model = BernoulliNB(fit_prior=True, alpha=np.float64(0.1))
model.fit(X_train, y_train)

# Predicción
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)

# Métricas
metrics = {"Accuracy Train: ": accuracy_score(y_train, y_pred_train),
           "Accuracy Test ": accuracy_score(y_test, y_pred_test)}

metrics
```

demasiado overfitting por tal motivo se intenta mejorar el modelo buscando mejores parametros

```python
hyperparams = {"alpha": np.linspace(0.8, 3.0, 4),
               "fit_prior": [True, False]}

# We initialize the random search
random_search = RandomizedSearchCV(model, hyperparams,
                                   n_iter=30,
                                   scoring="accuracy",
                                   cv=5,
                                   random_state=18)
random_search
random_search.fit(X_train, y_train)

random_search.best_params_
```

```python
# Entrenamiento del clasificador
model = BernoulliNB(fit_prior=False, alpha=np.float64(0.8))
model.fit(X_train, y_train)

# Predicción
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)

# Métricas
metrics = {"Accuracy Train: ": accuracy_score(y_train, y_pred_train),
           "Accuracy Test ": accuracy_score(y_test, y_pred_test)}

metrics
```
