# LECTURA DEL DATAFRAME

```python
df = pd.read_csv("https://raw.githubusercontent.com/4GeeksAcademy/naive-bayes-project-tutorial/main/playstore_reviews.csv")
```

package_name. Nombre de la aplicación móvil (categórico)

```python
df.head()
```

```python
df = df[["review","polarity"]]
```

```python
df["review"] = df["review"].str.strip().str.lower()
```

# SPLIT

```python
# Dividimos el conjunto de datos en muestras de train y test
X = df["review"]
y = df["polarity"]

# Instancia de: from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.2,
                                                    random_state=18)
```

## VECTORIZACION

```python
# Vectorización con TfidfVectorizer
vectorizer = TfidfVectorizer()

X_train = vectorizer.fit_transform(X_train).toarray()
X_test = vectorizer.transform(X_test).toarray()
```

# MODELO MULTINOMIAL

```python
# Entrenamiento del clasificador
model = MultinomialNB()

model.fit(X_train, y_train)
```

```python
X_train
```

```python
# Predicción
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)
```

```python
# Métricas
metrics = {"Accuracy Train: ": accuracy_score(y_train, y_pred_train),
           "Accuracy Test ": accuracy_score(y_test, y_pred_test)}

metrics
```

Observaciones:

se obtienen valores de test y train muy bajos por tal motivo se busca hacer una mejora de los hiperparametros del modelo

## MEJORA DE LOS HIPERPARAMETROS

```python
hyperparams = {"alpha": np.linspace(0.01, 10.0, 100),
               "fit_prior": [True, False]}

# We initialize the random search
random_search = RandomizedSearchCV(model, hyperparams,
                                   n_iter=100,
                                   scoring="accuracy",
                                   cv=5,
                                   random_state=18)
random_search
```

```python
random_search.fit(X_train, y_train)

random_search.best_params_
```

```python
# Entrenamiento del clasificador
model = MultinomialNB(fit_prior= True, alpha=np.float64(0.01))
model.fit(X_train, y_train)

# Predicción
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)

# Métricas
metrics = {"Accuracy Train: ": accuracy_score(y_train, y_pred_train),
           "Accuracy Test ": accuracy_score(y_test, y_pred_test)}

metrics
```

demasiado overfitting pero nos indica que el mejor numero esta por debajo de 1

```python
hyperparams = {"alpha": np.linspace(0.60, 0.70, 1),
               "fit_prior": [True, False]}

# We initialize the random search
random_search = RandomizedSearchCV(model, hyperparams,
                                   n_iter=50,
                                   scoring="accuracy",
                                   cv=5,
                                   random_state=18)
random_search
random_search.fit(X_train, y_train)

random_search.best_params_
```

Busqueda manual  del mejor alpha

```python
resultados = []

for alpha in np.arange(0.01, 1.01, 0.01):

    model = MultinomialNB(alpha=float(alpha))
    model.fit(X_train, y_train)

    y_train_pred = model.predict(X_train)
    y_test_pred  = model.predict(X_test)

    acc_train = accuracy_score(y_train, y_train_pred)
    acc_test  = accuracy_score(y_test, y_test_pred)
    diff = abs(acc_train - acc_test)

    resultados.append([alpha, acc_train, acc_test, diff])
    

# Convertimos a DataFrame para analizar
import pandas as pd
df_res = pd.DataFrame(resultados, columns=["alpha", "acc_train", "acc_test", "diff"])

# Selección del alpha óptimo:
#   1) maximize test accuracy
#   2) minimize difference train-test (para evitar overfitting)
df_res["score"] = df_res["acc_test"] - df_res["diff"]

best = df_res.loc[df_res["score"].idxmax()]
best
```

```python
xgb = XGBClassifier(n_estimators=50,
                    learning_rate=0.1,
                    random_state=18,
                    use_label_encoder=False,
                    eval_metric='mlogloss')
ada = AdaBoostClassifier(n_estimators=200,
                         learning_rate=0.1,
                         random_state=18)
gb = GradientBoostingClassifier(n_estimators=50,
                                learning_rate=0.1,
                                random_state=18)
lgb = LGBMClassifier(n_estimators=50,
                     learning_rate=0.1,
                     random_state=18)
```

```python
xgb.fit(X_train, y_train)
ada.fit(X_train, y_train)
gb.fit(X_train, y_train)
lgb.fit(X_train, y_train)
model_rf.fit(X_train, y_train)

# predicción
ada_y_pred_test = ada.predict(X_test)
gb_y_pred_test = gb.predict(X_test)
xgb_y_pred_test = xgb.predict(X_test)
lgb_y_pred_test = lgb.predict(X_test)
rf_y_pred_test = model_rf.predict(X_test)

ada_y_pred_train = ada.predict(X_train)
gb_y_pred_train = gb.predict(X_train)
xgb_y_pred_train = xgb.predict(X_train)
lgb_y_pred_train = lgb.predict(X_train)
model_rf_y_pred_train = model_rf.predict(X_train)

accuracy_train = (accuracy_score(y_train, ada_y_pred_train),
                  accuracy_score(y_train, gb_y_pred_train),
                  accuracy_score(y_train, xgb_y_pred_train),
                  accuracy_score(y_train, lgb_y_pred_train),
                  accuracy_score(y_train, model_rf_y_pred_train))

accuracy_test = (accuracy_score(y_test, ada_y_pred_test),
                 accuracy_score(y_test, gb_y_pred_test),
                 accuracy_score(y_test, xgb_y_pred_test),
                 accuracy_score(y_test, lgb_y_pred_test),
                 accuracy_score(y_test, rf_y_pred_test))

accuracy = pd.DataFrame(data=[accuracy_train, accuracy_test],
                        columns=['AdaBoost', 'Gradient Boosting', 'XGBoost', 'LightGBM','RandomFores'],
                        index=['Train set', 'Test set'])

accuracy
```
